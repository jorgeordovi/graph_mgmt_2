```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
x = [0,1,2,3,4]
y = [0,1,4,9,16]
y1 = [0,1,2,3,4]

plt.plot(x,y, label='Cuadrado')
plt.plot(x,y1, label='Linea')

plt.legend()
plt.title('Basic graph with Legend')
plt.show()
```

```python
plt.plot(x,y, label='Cuadrado')
plt.plot(x,y1, label='Linea')

plt.legend(loc='center left')
plt.title('Basic graph with center left Legend')
plt.show()
```

### Options Include

    - best(automatic)
    - upper right / left
    - lower right / left
    - right
    - center left / right
    - lower center
    - upper center
    - center

```python
plt.plot(x, y, label='Square',color='blue')
plt.plot(x, y1, label='Line', color='green')
plt.legend(loc='best', fontsize='medium', title='Graph Types', title_fontsize='large')
plt.title('Personanlized Graph ')
plt.show()
```

```python
### Higlight A specific value
plt.style.use('default')
plt.plot(x, y, label='Square',color='blue')
plt.plot(x, y1, label='Line', color='green')
plt.annotate('Unchained point', xy=(3,9), arrowprops=dict(facecolor='black', shrink=0.05))
plt.legend(loc='best', fontsize='medium', title='Graph Types', title_fontsize='large')
plt.title('Personanlized Graph ')
plt.show()
```

```python
### Higlight A specific value Capped

plt.style.use('default')
plt.plot(x, y, label='Square',color='blue')
plt.plot(x, y1, label='Line', color='green')

plt.annotate('Unchained point', xy=(3, 9), xytext=(2, 14), 
             arrowprops=dict(facecolor='red', shrink=0.05, width=2),
             fontsize=12, color='black', backgroundcolor='yellow')
plt.legend(loc='best', fontsize='medium', title='Graph Types', title_fontsize='large')
plt.title('Personanlized Graph ')
plt.show()
```

```python
### Advanced Notations

plt.style.use('default')
plt.plot(x, y, label='Square',color='blue')
plt.plot(x, y1, label='Line', color='orange')

for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', xy=(x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')
    
plt.title('Graph with multiple Notations')
plt.show()
```

```python
%matplotlib inline
plt.style.use('default')
plt.plot(x, y, label='Square',color='purple', linestyle='-.', marker='o')
plt.plot(x, y1, label='Line', color='green', linestyle='--', marker='s')

plt.annotate('Unchained point', xy=(3, 9), xytext=(2, 14), 
             arrowprops=dict(facecolor='red', shrink=0.05, width=2),
             fontsize=12, color='black', backgroundcolor='yellow')
plt.legend(loc='best', fontsize='large', title='Graph Types', title_fontsize='large')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid(True, color='orange')
plt.title('Personanlized Graph ')
plt.show()
```

```python
## 3d plots

from mpl_toolkits.mplot3d import Axes3D
#%matplotlib qt
%matplotlib inline
```

```python
x2 = np.linspace(-5,5,100)
y2 = np.linspace(-5,5,100)
X, Y = np.meshgrid(x2, y2)
Z = np.sin(np.sqrt(X**2 + Y**2))

#Create Axes
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X,Y,Z, cmap='viridis')

fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

ax.set_title('3D Surface Graph')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```
